## 冒泡排序

```python
DATA=[16,25,39,27,12,8,45,63]
```

```python
n=len(DATA)
print('原始顺序')
print(DATA)
for j in range(1,n-1):
    for i in range(n-j):
        if DATA[i]>DATA[i+1]:
            ptr=DATA[i]
            DATA[i]=DATA[i+1]
            DATA[i+1]=ptr
print('升序')            
print(DATA)
```

```python
#增加岗哨后的冒泡排序
n=len(DATA)
print('原始顺序')
print(DATA)
for j in range(1,n-1):
    #岗哨
    k=0
    for i in range(n-j):
     
        if DATA[i]>DATA[i+1]:
            ptr=DATA[i]
            DATA[i]=DATA[i+1]
            DATA[i+1]=ptr
            k+=1
    if k==0:
        break
print('升序')            
print(DATA)
```

## 插入排序

```python
#插入排序会造成大量的数据变动，所以用链表更为方便
class Node():
    def __init__(self):
        self.value=0
        self.next=None

#输入ptr，在ptr后插入数据
def insert(ptr,value):
    node=Node()
    node.value=value
    node.next=ptr.next
    ptr.next=node
    
#所有数据打印
def print_data(head):
    ptr=head.next
    while ptr.value!=10000:
        print(ptr.value,end=' ')
        ptr=ptr.next
        
DATA=[16,25,39,27,12,8,45,63]
```

```python
head=Node()
ptr=head
insert(ptr,10000)
for value in DATA:
    ptr=head
    while ptr.value<10000:
        if value <ptr.value:
            break
        else:
            pttr=ptr
            ptr=ptr.next
    insert(pttr,value)
```

## 希尔排序(感觉还可以优化)

将数据分为特定间隔的几个小块，以插入排序法排完区块内的数据后再渐渐减少间隔的距离

```python
class Node():
    def __init__(self):
        self.value=0
        self.next=None

#输入ptr，在ptr后插入数据
def insert(ptr,value):
    node=Node()
    node.value=value
    node.next=ptr.next
    ptr.next=node
    
    
def shell(DATA,time):
    print('第{}次数据：{}'.format(time,DATA))
    n=len(DATA)
    #步长
    step=int(n/(2**time))
    if step>0:
        #每组有length多个数，一共有step组
        length=int(n/step)
        head=[None]*step
        for j in range (step):
            head[j]=Node()
            ptr=head[j]
            insert(ptr,10000)
            data=[]
            for i in range(length):
                data.append(DATA[j+i*step])
            
            for value in data:
                ptr=head[j]
                while ptr.value<10000:
                    if value <ptr.value:
                        break
                    else:
                        pttr=ptr
                        ptr=ptr.next
                insert(pttr,value)
        #得到了head，还原成DATA的形式
        for i in range(step):
            ptr=head[i].next
            for j in range(length):
                DATA[j*step+i]=ptr.value
                ptr=ptr.next
        time=time+1
        shell(DATA,time)

```

## 快速排序

```python
#从左往右，寻找第一个比K大的数，若K最大，返回-1
def find_the_bigger(K,DATA):
    n=len(DATA)
    for i in range(n):
        if DATA[i]>K:
            return i
    return -1

#从右往左，寻找第一个比K小的数，若K最小，返回-1
def find_the_smaller(K,DATA):
    n=len(DATA)
    for i in range(n-1,-1,-1):
        if DATA[i]<K:
            return i
    return -1
def quick(DATA,left_index,right_index):
    #若左边界=有边界-1，则只有一个元素，不用排列
    if left_index<right_index: 
        #data为目标数据
        data=DATA[left_index:right_index+1]
        K=data[0]
        big_index=find_the_bigger(K,data)
        small_index=find_the_smaller(K,data)
        
        if big_index==-1:
            DATA[left_index],DATA[right_index]=DATA[right_index],DATA[left_index]
            right_index=right_index-1
            quick(DATA,left_index,right_index)
        elif small_index==-1:
            left_index+=1
            quick(DATA,left_index,right_index)
        else:
            #由于上面两个索引都是再data中的索引，所以还需要加上原来的开始边界恢复到DATA中的索引
            big_index=big_index+left_index
            small_index=small_index+left_index
            
            while big_index<small_index:
                #交换大数和小数的顺序
                DATA[big_index],DATA[small_index]=DATA[small_index],DATA[big_index]
                
                #重新检索大数和小数的索引
                data=DATA[left_index:right_index+1]
                big_index=find_the_bigger(K,data)
                small_index=find_the_smaller(K,data)
                big_index=big_index+left_index
                small_index=small_index+left_index
         
            DATA[left_index],DATA[small_index]=DATA[small_index],DATA[left_index]

            quick(DATA,left_index,small_index-1)
            quick(DATA,small_index+1,right_index)

    
    
```

```python

import time
import random
random.seed(1)
DATA=[]
for i in range(2000):
    DATA.append(random.randint(1,1000))
timeone=time.time()
quick(DATA,0,len(DATA)-1)
print(DATA)
timetwo=time.time()
print()
```

```python
print('所用时间为{}s'.format(timetwo-timeone))
```

## 基数排序法

```python
import time
import random
random.seed(1)
DATA=[]
for i in range(2000):
    DATA.append(random.randint(1,1000))
```

```python
#先将数据按个位数排列
a=time.time()
List=[[] for i in range(10)]
for data in DATA:
    for i in range(10):
        if data%10==i:
            List[i].append(data)
            break
new_list=[]
for i in range(10):
    new_list+=List[i]
    
#处理十位
List=[[] for i in range(10)]
for data in new_list:
    for i in range(10):
        if int(data/10)%10==i:
            List[i].append(data)
            break
new_list=[]
for i in range(10):
    new_list+=List[i]
    
#处理百位
List=[[] for i in range(10)]
for data in new_list:
    for i in range(10):
        if int(data/100)%10==i:
            List[i].append(data)
            break
new_list=[]
for i in range(10):
    new_list+=List[i]
    
#处理千位
List=[[] for i in range(10)]
for data in new_list:
    for i in range(10):
        if int(data/1000)%10==i:
            List[i].append(data)
            break
new_list=[]
for i in range(10):
    new_list+=List[i]
    
print(new_list)
b=time.time()
```

## 堆积排序

先建立最大堆积树，再删除根节点

```python
def compare(data,i):
    """
    将data[i]自下而上的比较，若大于父节点，则交换顺序
    data是列表，且索引从0开始
    i是目标数据的索引
    """
    if i>0:
        f=int((i-1)/2)
        if data[f]<data[i]:
            data[f],data[i]=data[i],data[f]
            compare(data,f)
```

```python
def build_tree(data):
    """
    生成最大堆积树
    """
    num=len(data)
    for i in range(1,num):
        compare(data,i)
    
```

```python
def heap(data):
    #最大索引值是num-1
    num=len(data)
    List=[0]*num
    for i in range(num):
        build_tree(data)
        value=data.pop(0)
        List[num-i-1]=value
    return List
```

```python
import time
import random
random.seed(1)
DATA=[]
for i in range(2000):
    DATA.append(random.randint(1,1000))
```

```python
a=time.time()
List=heap(DATA)
b=time.time()
print(List)
```

```python
print('用时{}s'.format(b-a))
```

# 合并排序

和之前的排序不同的是，合并排序是融合两个或多个已经排序好的序列

先比较两个排序第一个大小，再比较第二个，以此类推

```python
import random
random.seed(1)
List1=[]
List2=[]
for i in range(1000):
    List1.append(random.randint(1,9990))
List1.append(10000)
for j in range(1000):
    List2.append(random.randint(1,9990))
List2.append(10000)
```

先用选择排序将两个列表排序好

```python
n_1=len(List1)
for i in range(1,n_1):
    maxdata=0
    for j in range(n_1-i+1):
        if List1[j]>maxdata:
            maxdata=List1[j]
            maxindex=j
    List1[maxindex]=List1[n_1-i]
    List1[n_1-i]=maxdata
n_2=len(List2)
for i in range(1,n_2):
    maxdata=0
    for j in range(n_2-i+1):
        if List2[j]>maxdata:
            maxdata=List2[j]
            maxindex=j
    List2[maxindex]=List2[n_2-i]
    List2[n_2-i]=maxdata
```

```python
import time
a=time.time()
List=[]
l=0
r=0
for i in range(n_1+n_2-2):
    if List1[l]<List2[r]:
        List.append(List1[l])
        
        l+=1
        print('第{}个数是{}，来自列表1'.format(i+1,List1[l-1]))
    else:
        List.append(List2[r])
        
        r+=1
        print('第{}个数是{}，来自列表2'.format(i+1,List2[r-1]))
b=time.time()
print('用时{}s'.format(b-a))
```
